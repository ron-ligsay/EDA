```python
# Interquartile range (IQR)
Q1 = sales['Unit_Cost'].quantile(0.25)
Q3 = sales['Unit_Cost'].quantile(0.75)
IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# Normal distribution Formula
def normal_pdf(x, mu, sigma):
    return (1/ (sigma * np.sqrt(2 * np.pi))) * np.exp((-1 /2) * ((x - mu) / sigma) ** 2)
```

```python
normal_pdf(np.array([1, 2, 3]), 0, 1)
```

## Check normal distribution analytically
* kolmogorov-smirnov test
* shapiro-wilk test
* anderson-darling test


Null hypothesis - the data is normally distributed

where P-vlaue > 0.05, we fail to reject the null hypothesis (i.e. the data is normally distributed)

where P-value < 0.05, we reject the null hypothesis (i.e. the data is not normally distributed)

```python
import pandas as pd
import missingno as msno
```

```python
tree_census = pd.read_csv('../../datasets/2015StreetTreesCensus_TREES.csv')
tree_census_subset = tree_census[['tree_id','tree_dbh',
       'stump_diam', 'curb_loc', 'status', 'health', 'spc_latin', 'spc_common',
       'steward', 'guards', 'sidewalk', 'user_type', 'problems', 'root_stone',
       'root_grate', 'root_other', 'trnk_wire', 'trnk_light', 'trnk_other',
       'brnch_ligh', 'brnch_shoe', 'brnch_othe']]
tree_census_subset
```

```python
# Matrix Chart
msno.matrix(tree_census_subset)
```

```python
tree_census_subset.isnull().sum()
```

```python
# filter all the columns that have at least one missing value
tree_census_subset[tree_census_subset.isnull().any(axis=1)]
```

## Drop Missing Data

```python
# Drop rows containing NaN
tree_census_subset.dropna()
```

```python
# Drop columns containing NaN
tree_census_subset.dropna(axis=1)
```

## Fill the Nans

```python
# Fill wiuth a fixed value
tree_census_subset.fillna("No Observations") # or a number 
```

```python
# Fill Forward
tree_census_subset.fillna(method='ffill') # it fills the missing values with the previous value
```

```python
# Fill Backward
tree_census_subset.fillna(method='bfill') # it fills the missing values with the next value
```

```python
# Interpolate values
tree_census_subset.interpolate(method='linear') # it fills the missing values with the average of the previous and next value
```

```python
# Fill with mean
tree_census_subset.fillna(value=tree_census_subset.mean())
```

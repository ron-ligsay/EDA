```python
tree_census = pd.read_csv('../../datasets/2015StreetTreesCensus_TREES.csv')
tree_census
```

```python
tree_census_subset = tree_census[['tree_id','tree_dbh',
       'stump_diam', 'curb_loc', 'status', 'health', 'spc_latin', 'spc_common',
       'steward', 'guards', 'sidewalk', 'user_type', 'problems', 'root_stone',
       'root_grate', 'root_other', 'trnk_wire', 'trnk_light', 'trnk_other',
       'brnch_ligh', 'brnch_shoe', 'brnch_othe']]
tree_census_subset
```

```python
#tree_census_subset.fillna("Not Applicable")
mask = ((tree_census_subset['status'] == 'Stump') | (tree_census_subset['status'] == 'Dead'))
```

```python
tree_census_subset.loc[mask] = tree_census_subset.loc[mask].fillna('Not Applicable')
```

```python
tree_census_subset['problems'].fillna('None', inplace=True)
tree_census_subset['health'].fillna('Good', inplace=True)
tree_census_subset['spc_latin'].fillna('No Observation', inplace=True)
tree_census_subset['spc_common'].fillna('No Observation', inplace=True)
tree_census_subset['sidewalk'].fillna('NoDamage', inplace=True)
tree_census_subset['guards'].fillna('None', inplace=True)
```

```python
tree_census_subset.isna().sum()
```

```python
big_trees = tree_census_subset[(tree_census_subset['tree_dbh'] > 60) | (tree_census_subset['stump_diam'] > 60)]
big_trees
```

```python
tree_census_subsets = tree_census_subset[(tree_census_subset['tree_dbh'] <= 60) & (tree_census_subset['stump_diam'] <= 60)]
tree_census_subsets
```

```python
tree_census_subsets_alive = tree_census_subsets[(tree_census_subsets['status'] == 'Alive')]
tree_census_subsets_dead_or_stump = tree_census_subsets[(tree_census_subsets['status'] == 'Dead')|(tree_census_subsets['status'] == 'Stump')]
```

```python
tree_census_subsets_alive.groupby('spc_latin').mean()
```

```python
tree_census_subsets_alive.groupby('spc_latin')['tree_dbh'].describe()
```

```python
tree_census_subsets_alive[tree_census_subsets_alive['tree_dbh'] == 0]
```

```python
stats_alive = tree_census_subsets_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_alive
```

```python
tree_census_subsets_alive = tree_census_subsets_alive.merge(stats_alive, on='spc_latin', how='left')
tree_census_subsets_alive
```

```python
# when tree_dbh is less than 25% of the mean, replace with 25% of the mean
# when tree_dbh is greater than 75% of the mean, replace with 75% of the mean

tree_census_subsets_alive['tree_dbh'] = tree_census_subsets_alive.apply(lambda x: x['25%'] if x['tree_dbh'] < x['25%'] else x['tree_dbh'], axis=1)
tree_census_subsets_alive['tree_dbh'] = tree_census_subsets_alive.apply(lambda x: x['75%'] if x['tree_dbh'] > x['75%'] else x['tree_dbh'], axis=1)
```

```python
mask = tree_census_subsets_alive['tree_dbh'] < tree_census_subsets_alive['25%']
tree_census_subsets_alive.loc[mask, 'tree_dbh'] = tree_census_subsets_alive['25%']

mask = tree_census_subsets_alive['tree_dbh'] < tree_census_subsets_alive['75%']
tree_census_subsets_alive.loc[mask, 'tree_dbh'] = tree_census_subsets_alive['75%']
```
